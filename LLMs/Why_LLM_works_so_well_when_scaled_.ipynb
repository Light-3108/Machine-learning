As dimension increases, the probability of any 2 vectors to have angle close to 90 degree increases exponentially.

Dotproduct of any two vectors is close to zero means angle between them is 90 degree.

That means llm can store different view of the world in many directions.

It can have many different concepts.     If 2 vectors are 90 degree apart, It is likely that the 2 vector represents different concepts.

Hence each direction in the space have unique attribute or meaning helping LLM's to store and differentiate vast array of information.


```python
import numpy as np
```

```python
# x = np.arccos(0.5)
# print(np.degrees(x))
```

```python

random_vectors = np.random.uniform(-100,100, size = (1000,100))
```

```python
num_vectors = random_vectors.shape[0]
num_pairs = num_vectors * (num_vectors - 1) // 2

angles = np.zeros(num_pairs)
index = 0

for i in range(num_vectors):
    for j in range(i + 1, num_vectors):
        a = random_vectors[i]
        b = random_vectors[j]
        dot_product = np.dot(a, b)
        mag_a = np.linalg.norm(a)
        mag_b = np.linalg.norm(b)
        pre_angle = dot_product / (mag_a * mag_b)
        pre_angle = np.clip(pre_angle, -1.0, 1.0)
        pre_angle = np.arccos(pre_angle)
        angle = np.degrees(pre_angle)
        angles[index] = angle
        index += 1

angles_integers = np.floor(angles).astype(int)

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(angles_integers, bins=range(0, 180), edgecolor='black')
plt.title('Distribution of Angle Pairs')
plt.xlabel('Angle (degrees)')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(range(0, 180, 10))
plt.show()
```
